```python
#IMporting Libraries
import pandas as pd
import mysql.connector
```

```python
#database connection setup
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': 'root',
    'database': 'twitter_data'
}
```

```python
#Loading csv file
df = pd.read_csv(r"C:\Users\PRIYANKA\OneDrive\Desktop\task 2\twitter_data.csv")
```

```python
#Changing the column names according to the syntax of SQL
df = df.rename(columns={
    'Profile': 'profile',
    'Bio': 'bio',
    'Following Count': 'following_count',
    'Followers Count': 'followers_count',
    'Location': 'location',
    'Website': 'website'
})
```

```python
#Data Cleaning
```

```python
#Filling null values with N/A
df = df.fillna(value={'bio' : 'N/A', 'location':'N/A', 'website': 'N/A'})
```

```python
#Replacing hyphen with 0 in following_count column
df = df.fillna(value=0)
df['following_count'] = df['following_count'].replace('-', 0)
df
```

```python
# Remove commas and convert 'Following Count' column to integers, replacing NaN with 0
df['following_count'] = df['following_count'].str.replace(',', '')
df['following_count'] = df['following_count'].astype(float).fillna(0).astype(int)
df
```

```python
#Replacing hyphen with 0 in followers_count column
df['followers_count'] = df['followers_count'].replace('-', 0)
df
```

```python
# Convert 'Followers Count' column to numeric, handling 'M' and 'K' suffixes
#replaces 'M' with '*1e6' and 'K' with '*1e3'
df['followers_count'] = df['followers_count'].replace({'M': '*1e6', 'K': '*1e3'}, regex=True)
df['followers_count'] = pd.eval(df['followers_count'].replace('[^\d\.*]', '', regex=True))

# Round the values and convert to integers
df['followers_count'] = df['followers_count'].round().astype(int)
df
```

```python
# Connect to MySQL database
connection = mysql.connector.connect(**db_config)
cursor = connection.cursor()
```

```python
#Created a table named twitter_data and inserting the values into the table.
# Insert data into the table
for index, row in df.iterrows():
    insert_query = "INSERT INTO twitter_data(profile, bio, following_count, followers_count, location, website) VALUES (%s, %s, %s, %s, %s, %s);"
    cursor.execute(insert_query, (
        row['profile'],
        row['bio'],
        row['following_count'],
        row['followers_count'],
        row['location'],
        row['website']
    ))
```

```python
# Commit changes and close connection
connection.commit()
cursor.close()
connection.close()
```
